# Funciones basicas


```python
def greetUser(name):
    print(f"Hola, {name}!")
    
def calcMedia(lista):
    """Calculate the average of a list of numbers."""
    return sum(lista) / len(lista)
```

```python
greetUser("Albaro")

notas = [5,3,8,6,7,4]
calcMedia(notas)
```

# POO, clases

```python
class Car:
    # Contructor
    def __init__(self, make, model, year, km=0):
        """Initialize the Car object"""
        self.make = make
        self.model = model
        self.year = year
        self.km = km
    
    
    
    def drive(self, distance):
        """Simulate driving the car a certain distance"""
        self.km += distance
        print(f"Has conducido {distance} km. Total km: {self.km}")
    
    # Resumen de la información del coche
    def displayInfo(self):
        """Display the car's information"""
        name = f"{self.make} {self.model} {self.year}"
        print(name)
```

```python
coche1 = Car("Toyota", "Corolla", 2020, 0)
coche1.displayInfo()
coche1.drive(1500)
```

```python
coche1.drive(200)
```
