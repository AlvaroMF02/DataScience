# Ejercicio
- Calcular salario promedio del departamento
- Filtrar empleados con un salario mayor al promedio
- Guardar resultado en un csv

```python
# Create a dataset of employees
employees = {
    "Employee": ["John", "Jane", "Jim", "Jill", "Jack", "Jenny", "Joe", "Jessica", "James", "Julia"],
    "Department": ["HR", "Finance", "IT", "Marketing", "Sales", "HR", "Finance", "IT", "Marketing", "Sales"],
    "Salary": [50000, 60000, 55000, 70000, 65000, 52000, 61000, 56000, 71000, 66000]
}
df = pd.DataFrame(employees)
```

```python
df["Average_Salary_Department"] = df.groupby("Department")["Salary"].transform("mean")
```

```python
# filtrar empleados con salario por encima del promedio de su departamento
df[df["Salary"] > df["Average_Salary_Department"]]
```

```python
# guardar el dataframe en un csv
df[df["Salary"] > df["Average_Salary_Department"]].to_csv("high_salary_employees.csv", index=False)
```
