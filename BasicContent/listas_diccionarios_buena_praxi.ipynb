# Listas (array)

```python
estudiantes = ["pedro", "maria", "juan", "lucia", "carlos"]
```

```python
def printStudents(students):
    for i,stud in enumerate(students, start=1):
        print(f"Estudiante nº {i}: {stud}")
```

```python
printStudents(estudiantes)
```

# Diccionarios

```python
studenInfo = {
    "name": "Albaro",
    "age": 23,
    "city": "Madrid"
}   # KEY : VALUE

def printStudentInfo(listaEstudiantes):
    for key, value in listaEstudiantes.items(): # con enumerate() las keys se ven como numeros, con items() se ven las keys reales
        print(f"{key}: {value}")
    
```

```python
printStudentInfo(studenInfo)
```

# Ejercicio
- Vamos a hacer un programa que:
- Le pida al usuario cuántas veces quiere lanzar un dado.
- Genere un número aleatorio entre 1 y 6 por cada lanzamiento.
- Imprima los resultados.

```python
import random

def roll_dice(times):
    """ 
    Given an amount of times, it simulates rolling a dice that many times
    
    Args:
        times (int): Number of times the dice will be rolled
        
    Return:
        dice_results (list): List with the results of each roll
    """
    dice_results = []
    
    for _ in range(times):
        res = random.randint(1, 6)
        dice_results.append(res)
        
    return dice_results
```

```python
try:
    times = int(input("How many times do you want to roll the dice? "))
    if times < 1 or times > 50:
        raise ValueError()
except ValueError:
    print("Please enter a valid number (1 - 50).")
else:
    print(roll_dice(times))
```
